```python
import numpy as np
import matplotlib.pyplot as plt 

def makeGaussian(size, fwhm = 3, center=None):
    """ Make a square gaussian kernel.

    size is the length of a side of the square
    fwhm is full-width-half-maximum, which
    can be thought of as an effective radius.
    """

    x = np.arange(0, size, 1, float)
    y = x[:,np.newaxis]

    if center is None:
        x0 = y0 = size // 2
    else:
        x0 = center[0]
        y0 = center[1]

    return np.exp(-4*np.log(2) * ((x-x0)**2 + (y-y0)**2) / fwhm**2)

def Mask(frame2,sigma):
    [x1,x2,x3]=frame2.shape;
    g= makeGaussian(x1, sigma, [x1 ,x1/2]);
    g=np.delete(g, list(range(x1-x2)), 1)

    y = np.expand_dims(g, axis=3)
    y=np.repeat(y,3,axis=2)
    a = np.array(frame2, dtype=float)
    a = a*(1-y)
    img = a.astype(np.uint8)

    return(img)


g= makeGaussian(20, 10, [20 ,10]);
plt.imshow(Mask(frame2,60), cmap ="Greens")
plt.show() 
```
